```python
# %pip install pytrends
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
```

## Disable Warnings

```python
import warnings
warnings.filterwarnings("ignore")
warnings.simplefilter("ignore")
```

```python
trends = TrendReq()
```

```python
trends.build_payload(kw_list=["Stock Market"])
data = trends.interest_by_region()
data = data.sort_values(by="Stock Market", ascending=False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x = "geoName",
                        y = "Stock Market",
                        figsize = (15,10),
                        kind = "bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl = 'en-US', tz = 360)
data.build_payload(kw_list = ['Stock Market'])
data = data.interest_over_time()
fig, ax = plt.subplots( figsize = (15, 10) )
data['Stock Market'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for the Stock Market', fontweight = 'bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
